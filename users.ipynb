```python
class Person:
    def __init__(self, p_id, interests):
        self.p_id = p_id
        self.interests = interests
        
    def __str__(self):
        return 'id: {0} interests: {1}'.format(self.p_id, self.interests)
    
    def __repr__(self):
        return self.__str__()
        
def parse(p):
    i = 0
    num = ''
    if len(p) < 1:
        return None
    
    while p[i] != ' ':
        num = num + p[i]
        i += 1

    p_id = int(num)

    while p[i] != '[':
        i += 1

    s = i + 1

    while p[i] != ']':
        i += 1

    f = i

    interests = p[s:f].split(',')

    return Person(p_id, interests)
```

```python
with open('synthtetic_dataset.lg') as dataset_file:
    raw_data = dataset_file.read()
    
lines = raw_data.split('\n')

headers = lines[0:4]
body = lines[5:-1]
```

```python
persons = [parse(p) for p in body]
```
