```python
import requests as req
import numpy as np
```

```python
class Person:
    def __init__(self, p_id, interests):
        self.p_id = p_id
        self.interests = interests
        
    def __str__(self):
        return 'id: {0}\ninterests: {1}'.format(self.p_id, self.interests)
    
    def __repr__(self):
        return self.__str__()

def count_interests(node):
    count = 1
    if 'children' in node:
        for c in node['children']:
            count = count + count_interests(c)
    
    return count


interestsReq = req.get('https://gist.githubusercontent.com/jemsz95/1621a9cf70961ba797503bcae19af016/raw/8062d81488f11de88e67b8d6e38ee13a0a7d6afd/interests-tree-nested.json')
interests = interestsReq.json()

n_interests = 0


for interest in interests:
    n_interests += count_interests(interest)


def parse_interests(i_set, i_node, i_vec, i_p, weight):
    val = 0
    
    if 'children' in i_node:
        children = i_node['children']
        n_children = len(children)
        for c in children:
            val += parse_interests(i_set, c, i_vec, i_p, 1 / n_children)
            
    if i_node['name'] in i_set:
        val = 1
        
    i_vec[i_p[0]] = val * weight
    i_p[0] += 1
    
    return val


def parse(p):
    i = 0
    num = ''
    if len(p) < 1:
        return None
    
    while p[i] != ' ':
        num = num + p[i]
        i += 1

    p_id = int(num)

    while p[i] != '[':
        i += 1

    s = i + 1

    while p[i] != ']':
        i += 1

    f = i

    user_interests = frozenset(p[s:f].split(','))
    interests_vec = np.zeros(n_interests)
    i_p = [0]
    
    for interest in interests:
        parse_interests(user_interests, interest, interests_vec, i_p, 1 / len(interests))
    
    return Person(p_id, interests_vec)
```

```python
with open('synthtetic_dataset.lg') as dataset_file:
    raw_data = dataset_file.read()
    
lines = raw_data.split('\n')

headers = lines[0:4]
body = lines[5:-1]
```
